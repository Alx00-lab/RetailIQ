```python
import pandas as pd
from IPython.display import display

df_orders = pd.read_csv('brazilian-ecommerce/olist_orders_dataset.csv')
df_order_items = pd.read_csv('brazilian-ecommerce/olist_order_items_dataset.csv')
df_products = pd.read_csv('brazilian-ecommerce/olist_products_dataset.csv')
df_customers = pd.read_csv('brazilian-ecommerce/olist_customers_dataset.csv')
df_category_translation = pd.read_csv('brazilian-ecommerce/product_category_name_translation.csv')

display(df_orders)
display(df_order_items)
display(df_products)
display(df_customers)
display(df_category_translation)
```

```python
tables = {
    "orders": df_orders,
    "order_items": df_order_items,
    "products": df_products,
    "customers": df_customers,
    "category_translation": df_category_translation
}

for name, df in tables.items():
    print(f"\nTable: {name.upper()}")
    print("-" * 40)

    print("Shape:", df.shape)

    print("\nNull values:")
    nulls = df.isnull().sum()
    if nulls.sum() == 0:
        print("--- No nulls found ---")
    else:
        print(nulls[nulls > 0])

    print("\nData types:")
    print(df.dtypes)

    print("\nDuplicate rows:", df.duplicated().sum())
```

```python
missing_category = df_products['product_category_name'].isnull().sum()
print("Missing categories:", missing_category)
```

```python
df_products['product_category_name'] = df_products['product_category_name'].fillna('Other')
```

```python
df_orders['order_purchase_timestamp'] = pd.to_datetime(df_orders['order_purchase_timestamp'])
df_orders['order_delivered_customer_date'] = pd.to_datetime(df_orders['order_delivered_customer_date'])
df_orders[['order_delivered_customer_date']]
```

```python
zero_price = df_order_items[(df_order_items['price'] == 0) | (df_order_items['price'] < 0)]
print("Zero or negative price records:", len(zero_price))
```

```python
print(df_orders['order_status'].value_counts())
```

```python
df_orders['flag_missing_date'] = df_orders['order_delivered_customer_date'].isnull()
print("Missing values:", df_orders['flag_missing_date'].sum())
```

```python
invalid_delivered = df_orders[
    (df_orders['order_status'] == 'delivered') &
    (df_orders['order_delivered_customer_date'].isnull())
]
print("Delivered orders without delivery date:", len(invalid_delivered))
```

```python
invalid_shipped = df_orders[
    (df_orders['order_status'] == 'shipped') &
    (df_orders['order_delivered_customer_date'].isnull())
]
print("Delivered orders without shipped date:", len(invalid_shipped))
```

```python
print("Total customer_id:", df_customers['customer_id'].nunique())
print("Total customer_unique_id:", df_customers['customer_unique_id'].nunique())
```

```python
df_joined = df_order_items.merge(
    df_orders,
    on='order_id',
    how='inner'
)
print(df_joined.head())
```

```python
df_joined = df_joined.merge(
    df_customers[['customer_id', 'customer_unique_id']],
    on='customer_id',
    how='left'
)

df_joined = df_joined.drop(columns=['customer_unique_id_y'])
df_joined.columns
```

```python
df_joined = df_joined.drop(columns=['customer_unique_id_x'])
df_joined.columns
```

```python
df_joined['order_purchase_timestamp'] = pd.to_datetime(df_joined['order_purchase_timestamp'])
df_joined['order_delivered_customer_date'] = pd.to_datetime(df_joined['order_delivered_customer_date'])

df_joined['delivery_days'] = (
    df_joined['order_delivered_customer_date'] - 
    df_joined['order_purchase_timestamp']
).dt.days
```
